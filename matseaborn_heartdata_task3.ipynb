```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('heart_2022_no_nans.csv')
```

```python
df
```

# 1- STATE WISE PATIENTS DISTRIBUTION

```python
state_wise=df['State'].value_counts().head(10)
state_wise
```

```python
plt.plot(state_wise.index,state_wise.values,color='black',marker='*',markersize=10,markerfacecolor='r')
plt.xticks(rotation=90)
```

**INSIGHT:**
**It is visible that Washington state has most number of patients.**

# 2- SEX WISE PATIENTS DISTRIBUTIION

```python
plt.style.use
sex_wise=df['Sex'].value_counts()
plt.bar(sex_wise.index,sex_wise.values)
plt.xticks(rotation=90)
```

**INSIGHT:**
It can be derived that female patients are more in numbers compared to male patients.

# 3- HEALTH STATUS DISTRIBUTION OF TOTAL PATIENTS

```python
health_status=df['GeneralHealth'].value_counts()
health_status
```

```python
plt.stackplot(health_status.index,health_status.values, colors="g")
plt.xticks(rotation=90)
```

**INSIGHT:**
most of the patients are very good in general health.


#4-PATIENTS WITH KIDNEY DISEASE

```python
kidney_dis=df['HadHeartAttack'].value_counts()
plt.figure(figsize=(6,6))
wedgeprops={"edgecolor":"black","linewidth":3}
color=["pink","red"]
plt.pie(kidney_dis,labels=kidney_dis.index,wedgeprops=wedgeprops,autopct="%.1f%%",colors=color)
plt.title("PATIENTS WITH KIDNEY DISEASE")
plt.legend(bbox_to_anchor=(1,1))
```

**INSIGHT**
Only 5.5% of heart patients have kidney diseases

## **5- ALCOHOLIC HABITS OF MALE AND FEMALE HEART PATIENTS**

```python
plt.title("ALCOHOLIC HABITS OF MALE AND FEMALE HEART PATIENTS")
sns.countplot(x="Sex",data=df,hue='AlcoholDrinkers',palette="flare")
plt.legend(loc='upper center')
```

**INSIGHT:**
It is visible that both male and female patients are high in numbers in using alcohol as a habit among the heart patients

# 6- CATEGORISING THE HIGH RISK HEART PATIENTS WITH GENDER

```python
high_risk=df[(df['AlcoholDrinkers']=="Yes") & (df['HadHeartAttack']=="Yes") & (df['HadDiabetes']=="Yes") & (df['PhysicalActivities']=="No") & (df["BMI"]>30)]
```

```python
high_risky=high_risk['Sex'].value_counts()
high_risky
```

```python
plt.pie(high_risky,labels=high_risky.index,autopct="%.1f%%")
```

**INSIGHT:**
It is seen that male heart patients are more risky that of female patients

# 7- DISTRIBUTION OF SLEEPING HOURS AMONG HIGH RISK HEART PATIENTS

```python
sns.displot(high_risk['SleepHours'],kde=True,bins=6,color="pink")
```

**INSIGHT:**
Average sleeping hours of high risk heart patients is seen as 7 hours which is normal

# 8- DISTRIBUTION OF PHYSICAL HEALTH DAYS AMONG HIGH RISK HEART PATIENTS

```python

sns.displot(high_risk['PhysicalHealthDays'],kde=True,bins=6)
```

**INSIGHT:**
Average physical health days of high risk heart patients is 15 days

# 9- RELATION OF WEIGHT AND BMI OF HIGH RISK HEART PATIENTS

```python
sns.jointplot(x='WeightInKilograms',y='BMI',data=high_risk)
```

**INSIGHT:**
It can be interpretted that patients with average weight of 110 kg has an average BMI of 32.

# 10- RELATION OF HEIGHT AND BMI OF HIGH RISK HEART PATIENTS

```python
sns.jointplot(x='HeightInMeters',y='BMI',data=high_risk,kind='hex')

```

**INSIGHT:**
There are most number of patients with 1.8 meters of height with 31 as their BMI

# 11- DISTRIBUTION OF HEART PATIENTS AMONG THE AGE GROUP 18-24

```python
AgeCat=df.groupby("AgeCategory")
Low_Age_group=AgeCat.get_group("Age 18 to 24")
low_group=Low_Age_group['HadHeartAttack'].value_counts()
low_group
```

```python
plt.title(" DISTRIBUTION OF HEART PATIENTS AMONG THE AGE GROUP 18-24")
color=["pink","red"]
plt.pie(low_group,labels=low_group.index,colors=color,autopct="%.1f%%")
plt.legend(bbox_to_anchor=(1,1))
```

**INSIGHT:**
It is seen that only 0.4% of patients are under the age group 18-24

# 12- AGE DISTRIBUTION OF HIGH RISK HEART PATIENTS

```python
age_dist=high_risk['AgeCategory'].value_counts()
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(10,6))
plt.title("AGE DISTRIBUTION OF HIGH HEART PATIENTS")
plt.bar(age_dist.index,age_dist.values,color="brown")
plt.xlabel("AgeCategory")
plt.ylabel("No.of Heart patients")
plt.xticks(rotation=90)
plt.legend()
```

**INSIGHT:**
It is visible that patients of age group 70-74 has most number of high risk heart patients

# 13- DENSITY OF BMI OF HIGH RISK HEART PATIENTS

```python
plt.style.use('dark_background')
sns.rugplot(high_risk['BMI'])
sns.kdeplot(high_risk['BMI'])
```

**INSIGHT:**
Density of BMI of high risk heart patients is in between 30-40

# 14- AVERAGE WEIGHT OF MALE AND FEMALE HEART PATIENTS

```python
plt.style.use('dark_background')
sns.barplot(x='Sex',y='WeightInKilograms',data=df,estimator=np.mean,color='pink')
plt.title('AVERAGE WEIGHT OF MALE AND FEMALE')
```

**INSIGHT:**
Average weight of male patients= 90kg
Average weight of female patients= 75kg

# 15- GENERAL HEALTH STATUS AND SLEEPING HOURS OF HIGH RISK HEART PATIENTS

```python
table=high_risk.sort_values(by='SleepHours',ascending=True)
```

```python
sns.lineplot(x='GeneralHealth',y='SleepHours',data=table,errorbar=None,color='cyan',linewidth=2,marker='*',markersize=10,markerfacecolor='red')
```

# 16- GENERAL HEALTH AND BMI STATUS OF HIGH RISK HEART PATIENTS WITH GENDER

```python
sns.lineplot(x='GeneralHealth',y='BMI',data=high_risk,hue='Sex',errorbar=None,color='cyan',linewidth=2,marker='*',markersize=10,markerfacecolor='red',estimator=np.max,palette='flare')
```

# 17- HOW MANY MALE AND FEMALE HEART PATIENTS HAVE HAD STROKE

```python
sns.barplot(x='Sex',y='BMI',data=df,hue='HadStroke',estimator=np.std,palette='pastel')
```

# 18- STATE WISE AGE DISTRIBUTION OF TOTAL HEARRT PATIENTS

```python
sns.countplot(x='AgeCategory',data=high_risk,color='blue')
plt.xticks(rotation=90)
```

# 19- STATE WISE GENDER DISTRIBUTION OF HIGH RISK HEART PATIENTS

```python
sns.countplot(x='State',hue='Sex',data=high_risk)
plt.xticks(rotation=90)
```

# 20- BOX PLOTING OF SLEEPING HOURS AND GENERAL HEALTH STATUS OF HIGH RISK HEART PATIENTS
