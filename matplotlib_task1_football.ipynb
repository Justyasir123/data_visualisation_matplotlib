#  1- TOP 10 BEST RATED PLAYERS

```python
top_10_players=df.sort_values(by='Overall',ascending=False).head(10)
```

```python
plt.style.use('classic')
plt.bar(top_10_players['Name'],top_10_players['Overall'],color='r')
plt.title('TOP 10 PLAYERS AND THEIR RATING')
plt.xlabel('PLAYERS')
plt.ylabel('RATING')
plt.ylim(88,95)
plt.xticks(rotation=90)
```

**INSIGHT:**
L.Messi and Critiano Ronaldo are the two top rated players in today's football era.

# 6- CLUB DISTRIBUTION BASED ON THEIR VALUE

```python
def convert_value(value):
  if value.endswith('M'):
    return float(value[1:-1])*1000000
  elif value.endswith('K'):
    return float(value[1:-1])*1000
  else:
    return float(value[1:])
```

```python
df['Value'] = [convert_value(value) for value in df['Value']]
```

```python
High_valued_club=df.groupby("Club")["Value"].sum().sort_values(ascending=False).head(10)
High_valued_club
```

```python
plt.pie(High_valued_club,labels=High_valued_club.index,autopct="%1.1f%%")
plt.title("Club distribution based on their value")
plt.legend(loc="upper right",bbox_to_anchor=(1.5,1))
```

**INSIGHT**
Real Madrid is having the highest valued club based on sum of players value
