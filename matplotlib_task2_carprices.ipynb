```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('car_prices.csv')
```

```python
df
```

```python
df.columns
```

```python
df.shape
```

# 1- YEAR WISE SALES REPORT

```python
year_wise_sales=df['year'].value_counts()
year_wise_sales
```

```python
plt.plot(year_wise_sales.index,year_wise_sales.values,color='black',marker='*',markersize=10,markerfacecolor='r')
plt.xlabel('sales quantity')
plt.ylabel('YEAR')
plt.title('YEAR WISE SALES REPORT')
```

**INSIGHT:**
It is seen that 2012 is the highest sales year with over 42k units sold

# 2- BRAND WISE SALES REPORT

```python
brand_wise_sales=df['make'].value_counts().head(10)
brand_wise_sales
```

```python
plt.style.use( 'dark_background')
brand_wise_sales.plot(kind='bar',x=brand_wise_sales.index,y=brand_wise_sales.values,color='b')
plt.title('BRAND WISE SALES REPORT')
plt.xlabel('BRANDS')
plt.ylabel('QUANTITY')
```

**INSIGHT:**
Ford has achieved high sales among others with over 40000 units sold.

# 3- USERS OF MANUAL AND AUTOMATIC TRANSMISSIONS

```python
gear_group=df['transmission'].value_counts().head(10)
```

```python
wedgeprops={'edgecolor':'black','linewidth':5}
plt.pie(gear_group,labels=gear_group.index,colors=['r','b'],wedgeprops=wedgeprops,autopct='%1.1f%%')
plt.title('TRANSMISSION DISTRIBUTION')
```

**INSIGHT:**
Users of Automatic transmission are far high (96.4%) than of manual transmission users

#4- RELATION BETWEEN KILOMETERS COMPLETED AND THEIR SELLING PRICES

```python

plt.scatter(df['odometer'],df['sellingprice'],color='r')
plt.title('RELATION BETWEEN KILOMETERS COMPLETED AND THEIR SELLING PRICES')
plt.xlabel('KILOMETERS')
plt.ylabel('SELLING PRICE')
```

**INSIGHT:**
Negative relationship is seen in between kilometers and selling price

# 5- TOP SELLING MODELS OF KIA COMPANY BASED ON SELLING PRICE

```python
filt1= df['make']=='kia'
kiafilter=df[filt1]
kiafilter
```

```python
kiafilter['model'].unique()
```

```python
varient=kiafilter.groupby('model')['sellingprice'].mean()
varient
```

```python
plt.bar(varient.index,varient.values)
plt.title('TOP SELLING MODELS BASED ON SELLING PRICE')
plt.xlabel('KIA VARIENTS')
plt.ylabel('SELLING PRICE')
```

**INSIGHT:**
Borrego has the top selling rate among other models of KIA

# 6- CARS WITH HIGH SELLING PRICE

```python
high_selling_price=df.sort_values(by="sellingprice",ascending=False).head(10)
```

```python
plt.bar(high_selling_price["make"],high_selling_price["sellingprice"],color="pink")
plt.title("Cars with high selling price")
plt.xlabel("cars")
plt.ylabel("selling price")
```

**INSIGHT:**
Ford is experiencing highest sales followed by Ferrari.

# 7- LOWEST MARKET VALUE OF DIFFERENT BRANDS

```python
low_mmr=df.groupby("make")["mmr"].sum().sort_values(ascending=False).tail(10)
```

```python
plt.style.use("dark_background")
low_mmr.plot(kind="line",x=low_mmr.index,y=low_mmr.values,color="brown",lw=5)
plt.xticks(rotation=90)
plt.title("brands having lowest marketing price")
plt.ylabel("marketing price")
```

**INSIGHT:**
Honda is experiencing lowest car price

# 8- BRANDS HAVING HIGHEST MARKET PRICE

```python
high_mmr=df.groupby("make")["mmr"].sum().sort_values(ascending=False).head(10)
```

```python
plt.style.use("classic")
high_mmr.plot(kind="bar",x=high_mmr.index,y=high_mmr.values,color="red")
plt.title("Brands with highest market price")
plt.xlabel("BRANDS")
plt.ylabel("MARKET PRICE")
```

**INSIGHT:**
Ford is having highest market selling price

# 9- CARS WITH HIGH MARKET VALUE BASED ON COLORS WHITE & GREY

```python
filt3=(df['color']=='white') | (df['color']=='gray')
filter4=df[filt3]
mmr_base=filter4.sort_values(by='mmr',ascending=False)
color_base=mmr_base["color"].value_counts()
plt.pie(color_base,labels=color_base.index,autopct='%1.1f%%')
plt.title('WHITE & GREY MARKET VALUE DISTRIBUTION')
```

**INSIGHT:**
It can be understood that white color cars have more market value than of grey cars


# 10- TOP 10 CAR BRANDS WITH BEST CONDITION

```python
condition_vehicles=df.sort_values(by="condition",ascending=False).head(10)
condition_vehicles_name=condition_vehicles["make"].head(10)
Top_condition_vehicles=condition_vehicles["condition"]
plt.bar(condition_vehicles_name,Top_condition_vehicles,color="Red")
plt.xticks(rotation=90)
plt.xlabel("Car Brand")
plt.ylabel("Condition Rating")
plt.title("TOP 10 BRANDS WITH BEST CONDITION")
```

**INSIGHT:**
ALL CARS HAVE EQUAL CAR CONDITION

# 11- CAR BRAND WITH HIGHEST MARKET VALUE

```python
plt.style.use('dark_background')
year_group=df.groupby("year")
market_value_2015=year_group.get_group(2015).sort_values(by="mmr",ascending=False)
market_value_cars_2015=market_value_2015["make"].head(10)
top_market_value_2015=market_value_2015["mmr"].head(10)
plt.plot(market_value_cars_2015,top_market_value_2015,"g*-",markerfacecolor="red",markersize=15)
plt.xlabel("Car Brand")
plt.ylabel("Market Value")
```

**INSIGHT:**
Mercedes Benz has the highest market value than BMW.

# 12- CARS MANUFACTURED IN 2014  AND THEIR COLORS

```python
cars_2014=year_group.get_group(2014).head(10)
car_colors_2014=cars_2014["color"]
car_names_2014=cars_2014["make"]
plt.scatter(car_names_2014,car_colors_2014)
plt.title("2014 manufacturing card and their colors")
plt.xlabel("Car Brand")
plt.ylabel("Car colors")
```

**INSIGHT:**
It can be understood that chevrolet brand has more color options like gray,black, red and blue

# 13- 2015 MANUFACTURED CARS AND THEIR PRICES

```python
cars_2015=year_group.get_group(2015).sort_values(by="sellingprice",ascending=False)
Top_sellingprice_cars_2015=cars_2015["sellingprice"].head(10)
Top_sellingprice_car_names_2015=cars_2015["make"].head(10)
plt.bar(Top_sellingprice_car_names_2015,Top_sellingprice_cars_2015,color="green")
plt.title("2015 manufactured cars and their price")
plt.xlabel("Car Brand")
plt.ylabel("Selling Price")
```

**INSIGHT**
Mercedes Benz has highest selling price than BMW.

# 14- TOTAL COLORS TOYOTA HAS LAUNCHED

```python
car_group=df.groupby("make")
toyota_cars=car_group.get_group("Toyota").head(10)
toyota_car_colors=toyota_cars["color"].value_counts()
plt.pie(toyota_car_colors,labels=toyota_car_colors.index,autopct="%1.1f%%")
plt.title("Toyota car colors")
plt.legend(bbox_to_anchor=(1.5,1))
```

#15- LEXUS BRAND'S TOP SELLING INTERIOR COLORS

# 18-Top 10 sedan selling brands

```python
plt.style.use('dark_background')
filtersedan=df[df['body']=='Sedan']
sedan_brands=filtersedan['make'].value_counts().head(10)
plt.bar(sedan_brands.index,sedan_brands.values)
plt.xticks(rotation=90)
plt.xlabel('BRANDS')
plt.ylabel('SALES')
plt.title('Top 10 sedan selling brands')
```

# 19- SATE WISE SALES ANALYSIS IN AMERICA

```python
state=df.groupby("state")["sellingprice"].sum().sort_values(ascending=False).head(10)
state
plt.style.use("dark_background")
plt.figure(figsize=(8,8))
plt.pie(state,labels=state.index,autopct="%1.1f%%")
plt.title("STATE WISE SALE ANALYSIS")
plt.legend(bbox_to_anchor=(1,1))
```

**INSIGHT:**
California is having highest sales among other states with 18.4%
